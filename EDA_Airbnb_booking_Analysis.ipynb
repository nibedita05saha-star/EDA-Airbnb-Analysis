```python
#average price of listing across neighbourhood groups
```

```python
avg_price = df.groupby('neighbourhood_group')['price'].mean().reset_index()
print(avg_price)
```

```python
#Using horizontol bar chart to show average price across neighbourhood groups
import matplotlib.pyplot as plt

avg_price = df.groupby('neighbourhood_group')['price'].mean().sort_values()

avg_price.plot(kind='barh', figsize=(8,5), color='pink')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood Group')
plt.title('Average Airbnb Price by Neighbourhood Group')
plt.show()
```

```python
#How does price vary across room types?
```

```python
df.groupby('room_type')['price'].mean()
```

```python
#Which neighbourhoods have the highest/lowest average prices?
```

```python
# Average price per neighbourhood
avg_prices = df.groupby("neighbourhood_group")["price"].mean()

# Neighbourhood with highest average price
highest = avg_prices.idxmax(), avg_prices.max()

# Neighbourhood with lowest average price
lowest = avg_prices.idxmin(), avg_prices.min()

print("Highest average price:", highest)
print("Lowest average price:", lowest)
```

```python
#What is the average preferred price by customers according to the location?
```

```python

avg_preffered_price_df = df.groupby(['neighbourhood_group','room_type'])['price'].mean().unstack()
avg_preffered_price_df
```

```python
#Graph of average price of property according to location using grouped bar chart
avg_preffered_price_df.plot.bar()
```

**2. Location-Based Insights**

Which neighbourhood group has the most listings?

Show number of listings in each neighbourhood group.


```python
#neighbourhood with most listings
result = df['neighbourhood_group'].value_counts().idxmax()
print(result)
```

```python
#Show number of listings in each neighbourhood group.
no_of_host_per_location = df.groupby('neighbourhood_group')['host_id'].count()

print(no_of_host_per_location)
```

```python
plt.legend(labels = ['No. of Host'])
plt.plot(no_of_host_per_location)
plt.title('Number of host per location')
plt.ylabel('Host')
plt.xlabel('Location')
```

**3. Host Insights**

What are the top 10 listings as per neighbourhood groups?





```python
#How many listings does each host have?

# Count listings per host
host_counts = df.groupby('host_id')['id'].count().reset_index(name='listing_count')

# Get unique host information: host_id, host_name, neighbourhood_group
host_info = df[['host_id', 'host_name', 'neighbourhood_group']].drop_duplicates(subset='host_id')

# Combine listing counts with host details
host_summary = host_counts.merge(host_info, on='host_id')

# Get top 10 hosts with the highest number of listings
top10_hosts = host_summary.sort_values('listing_count', ascending=False).head(10)

print(top10_hosts)
```

**4. Review Insights**

Which neighbourhood has the highest average number of reviews?

How does reviews_per_month vary across neighbourhood groups?


```python
df.groupby('neighbourhood_group')['number_of_reviews'].mean().sort_values(ascending=False).head(1)
```

```python
df.groupby('neighbourhood_group')['reviews_per_month'].mean().sort_values(ascending=False)
```

Here we see Staten Island though has highest reviews, average monthly reviews are more in Queens

```python
grouped = df.groupby('neighbourhood_group')['reviews_per_month'].mean().sort_values()
plt.figure(figsize=(8,5))
plt.scatter(grouped.values, grouped.index)
plt.title("Average Reviews per Month by Neighbourhood Group")
plt.xlabel("Reviews per Month")
```

**5. Availability Insights**


Which room types are the least/most available?


```python
df.groupby('room_type')['availability_365'].mean().sort_values()
```

```python
import matplotlib.pyplot as plt

# Compute the grouped mean values
availability_by_room = (
    df.groupby('room_type')['availability_365']
      .mean()['/[`  ]']
      .sort_values()
)

# Custom colors (one per room type)
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999']

# Plot pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    availability_by_room,
    labels=availability_by_room.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Average Availability (365 days) by Room Type')
plt.tight_layout()
plt.show()
```
